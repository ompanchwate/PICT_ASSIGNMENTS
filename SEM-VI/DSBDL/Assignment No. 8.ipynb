```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
```

###**Loading the Train Data**###

```python
training_data = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Datasets/train.csv")
```

###**Loading the Test Data**###

```python
testing_data = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Datasets/test.csv")
```

```python
testing_data.describe()
```

```python
testing_data.head()
```

###**Linear Regression Model to predict Home Prices**###

```python
X_train = training_data[['ID', 'crim', 'zn', 'indus', 'chas', 'nox', 'rm', 'age', 'dis', 'rad', 'tax', 'ptratio', 'black', 'lstat']]
y_train = training_data['medv']
```

```python
X_test = testing_data[['ID', 'crim', 'zn', 'indus', 'chas', 'nox', 'rm', 'age', 'dis', 'rad', 'tax', 'ptratio', 'black', 'lstat']]
y_test_pred = np.zeros(len(testing_data))
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
print("The Predicted Values of 'medv' Column:\n", *y_pred,sep='\n')
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mse = mean_squared_error(y_test_pred, y_pred)
print(f"Mean Squared Error: {mse}")
```
