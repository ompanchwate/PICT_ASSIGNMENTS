#**1. Data Wrangling - I**#

```python
import pandas as pd
```

###**Dataset - 1: Iris Dataset**###

```python
csvfile1 = "/content/IRIS.csv"
```

###**Load the Dataset into Pandas DataFrame**###

```python
#Loading the Dataset into 'data' variable
data = pd.read_csv(csvfile1, encoding='latin-1')
```

```python
display(data)
```

```python
print(data.columns)
```

###**Displaying Null Values**###

```python
#Displaying the Null Values in the DataFrame
display(data.isnull())
```

```python
display(data.isnull().sum())
```

```python
#describe() Function
data.describe()
```

```python
#Printing the Datatypes of the Variables
print(data.dtypes)
```

```python
#Dimensions of the DataFrame
rows, columns = data.shape
print(f"Number of Rows: ", rows)
print(f"Number of Columns: ", columns)
```

###**Convert Categorical Variables into Quantitative Variables**###

```python
#Converting Categorical Variables into Quantitative Varibles
df_encoded = pd.get_dummies(data, columns=['species'])
display(df_encoded)
```
